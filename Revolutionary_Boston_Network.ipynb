```r
#Access required library
library('igraph')
```

```r
#Read in data
data <- as.matrix(read.csv("http://shroutdocs.org/dcs375/Intel.csv", row.names = 1))
```

```r
data
```

```r
#The person net shows the number of organizations that people share, numbers in columns and row overlaps show how many groups
#they share
person.net <- data %*% t(data)

#The group net shows us how mny people were within both the overlapping groups, numbers in columns and row overlaps show 
#how many people shared those groups
group.net <- t(data) %*% data

#Create the diagonal with NA within every position we do this because you cannot share a group with yourself
diag(group.net) <- NA
diag(person.net) <- NA
```

```r
person.net
```

```r
#Person.g shows users all the relationships between people who were within the same group, each line has 2 names 
#shared by a organization
person.g <- graph.adjacency(person.net, mode = "undirected", weighted=NULL, diag=FALSE)

#Group.g shows users all the groups that had members who were within both, each line has 
#2 groups that were shared by a person 
group.g <- graph.adjacency(group.net, weighted = TRUE, mode = "undirected", diag=FALSE)

#la produces vertices (7 terrorist groups) on the plane to show the potential location of each nodes
la <- layout_with_fr(group.g)

#E.wt provides us with the list of all the edge attributes, the number of edges and the weights
e.wt <- edge_attr(group.g, "weight")
```

```r
#This plot shows us all the potential groups and the visual connections between them where the thicker the line is,
#the more overlap that occured.
plot(group.g, layout=la, vertex.size=15, edge.width=e.wt,vertex.label=V(group.g)$name)
```

From this plot I learned that there were a lot more overlap of people being in the same group between the Tea Party, London Enemies, Boston Committee and North Caucus. This means that these organizations typically had people within all three.

```r
#la2 produces all the vertices of all the potential people within this network
la2 <- layout.fruchterman.reingold(person.g)

#e.wt2 produces all the edges connecting people who have a shared organzation group
e.wt2 <- get.edge.attribute(person.g, "weight")

#This plot shows a visualization of the connections between people based off of the organizations that they shared.
plot(person.g, layout = la2, vertex.size = 3, edge.width=0.1, vertex.label =V(person.g)$name)
```

From the plot above we can she how the network enabled us to see the connections between certain people based on the organizations in which they shared. One obvious observation that can be made is that there was no one person within their organization by themself. 

```r
##Betweenness

#This line calculates the betweenness of each person within our initial data set
btwn.person <- betweenness(person.g)

#This line assigns the corresponding name with its corresponding betweenness value
names(btwn.person) <- V(person.g)$name

#This line now rearrages the data in decreasing order of betweennees value
ind <- order(-btwn.person)

#This line allow us to see the top ten highest people with betweenness values 
btwn.person[ind][1:10]

#The betweenness value is a value that allows us to see how often a person can be used as a link or gateway between other
# people within the network
```

These betweenness scores tells us how many times a person could be used to relate to people who otherwise would not have any direct connecting edges.

```r
## Eigenvector 

#This line calculates the centrality of each person within our initial data set
cent.eig <- evcent(person.g)

#This line assigns the corresponding name with its corresponding centrality value
names(cent.eig$vector) <- V(person.g)$name

#This line now rearrages the data in decreasing order of centrality value
ind <- order(-cent.eig$vector)

#This line allow us to see the top ten highest people with centrality values 
cent.eig$vector[ind][1:10]
```

These centrality scores tell us a proportional value to how much a person is used to connect people who would otherwise not have any connecting edges.
